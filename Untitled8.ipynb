```python
import pandas as pd

train_df = pd.read_csv("/content/train.csv", on_bad_lines="skip", engine="python")
test_df = pd.read_csv("/content/test.csv", on_bad_lines="skip", engine="python")

print(train_df.head())
```

```python
train_df["text"] = train_df["Title"].astype(str) + " " + train_df["Description"].astype(str)
test_df["text"] = test_df["Title"].astype(str) + " " + test_df["Description"].astype(str)

print(train_df[["text"]].head())
```

```python
train_df = train_df[["text", "Class Index"]]
test_df = test_df[["text", "Class Index"]]

train_df = train_df.rename(columns={"Class Index": "label"})
test_df = test_df.rename(columns={"Class Index": "label"})
```

```python
from datasets import Dataset
from transformers import AutoTokenizer

# 1. Reset datasets from the original pandas DataFrames
train_dataset = Dataset.from_pandas(train_df)
test_dataset = Dataset.from_pandas(test_df)

# 2. Define the shift (Original 1-4 -> Target 0-3)
def shift_labels(example):
    example["label"] = example["label"] - 1
    return example

train_dataset = train_dataset.map(shift_labels)
test_dataset = test_dataset.map(shift_labels)

# 3. Define and Apply Tokenization
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")

def tokenize(batch):
    return tokenizer(
        batch["text"],
        truncation=True,
        padding="max_length",
        max_length=256
    )

train_dataset = train_dataset.map(tokenize, batched=True)
test_dataset = test_dataset.map(tokenize, batched=True)

# 4. Final formatting for Trainer
train_dataset = train_dataset.remove_columns(["text"])
test_dataset = test_dataset.remove_columns(["text"])
train_dataset.set_format("torch")
test_dataset.set_format("torch")
train_dataset = train_dataset.rename_column("label", "labels")
test_dataset = test_dataset.rename_column("label", "labels")

# Verify labels
unique_labels = set([int(x) for x in train_dataset["labels"]])
print(f"Unique labels in train: {unique_labels}")
```

### Test du modèle avec des exemples personnalisés
Nous utilisons la `pipeline` de Transformers pour simplifier l'inférence.

```python
import torch
from transformers import pipeline

# Créer un pipeline de classification
classifier = pipeline("text-classification", model=model, tokenizer=tokenizer, device=0 if torch.cuda.is_available() else -1)

# Liste de classes pour référence (AG News dataset)
class_names = ["World", "Sports", "Business", "Sci/Tech"]

# Exemples à tester
custom_texts = [
    "The stock market saw a significant rise today as technology shares surged.",
    "The national team won the championship after a thrilling final match.",
    "New discoveries in space exploration reveal water on distant planets.",
    "Peace talks have resumed between the two neighboring countries."
]

# Prédiction
results = classifier(custom_texts)

for text, result in zip(custom_texts, results):
    label_idx = int(result['label'].split('_')[-1])
    print(f"Texte: {text}")
    print(f"Prédiction: {class_names[label_idx]} (Score: {result['score']:.4f})\n")
```

### Sauvegarde et Téléchargement du Modèle
Nous sauvegardons le modèle et le tokenizer, puis nous compressons le dossier pour faciliter le téléchargement sur votre ordinateur.

```python
import os
import shutil
from google.colab import files

# 1. Sauvegarde locale dans Colab
save_path = "./my_news_model"
model.save_pretrained(save_path)
tokenizer.save_pretrained(save_path)

# 2. Compression en fichier ZIP
shutil.make_archive("news_model_export", 'zip', save_path)

# 3. Téléchargement vers votre ordinateur
files.download("news_model_export.zip")

print("Le modèle a été sauvegardé et le téléchargement du fichier .zip va commencer.")
```
